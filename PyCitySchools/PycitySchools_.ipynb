# PyCity Schools Analysis



```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Files to Load 
school_data_path = "Resources/schools_complete.csv"
student_data_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

# District Summary

```python
# Cell to calculate the following totals, Averages, percentages and then create a data frame. 

school_count = len(school_data_complete["school_name"].unique())
student_count = len(school_data_complete["Student ID"].unique())
total_budget = school_data["budget"].sum()
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()


passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage

passing_reading_count = (school_data_complete[school_data_complete["reading_score"] >= 70]["student_name"].count())/student_count * 100
passing_reading_count

overall_passing_rate = ((school_data_complete.loc[(school_data_complete["math_score"]>=70)&
                                                  (school_data_complete["reading_score"]>=70),
                                                  "Student ID"].count())/student_count)*100

district_summary = pd.DataFrame({"Total Schools": school_count,
                                "Total Students": student_count,
                                "Total Budget": total_budget,
                                "Average Math Score": average_math_score,
                                "Average Reading Score": average_reading_score,
                                "% Passing Math": passing_math_count,
                                "% Passing Reading": passing_reading_count,
                                "% Overall Passing Rate": [overall_passing_rate]})
district_summary
```

```python
district_summary

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame with applied formatting
district_summary
```

School Summary

```python
# Dataframe for the school summary

school_summary_df = school_data_complete[["student_name","school_name","reading_score","math_score","type","size","budget"]]

# setting the school name as a index

School_type = school_data.set_index(["school_name"])["type"]

# total number of students per school

student_count = school_summary_df["school_name"].value_counts()

# total school budget

school_budget = school_summary_df.groupby(["school_name"])["budget"].mean()

# total per student 

student_budget = school_budget/student_count

# Ave math score per school 

average_math = school_summary_df.groupby(["school_name"])["math_score"].mean()

# Ave reading score per school 

average_reading = school_summary_df.groupby(["school_name"])["reading_score"].mean()

# The number of students per school with math scores of 70 or higher

Total_passing_math = school_summary_df.loc[school_summary_df["math_score"] >= 70].groupby("school_name").count()["student_name"]

percentage_math = (Total_passing_math/student_count)*100

# The number of students per school with reading scores of 70 or higher

Total_passing_reading = school_summary_df.loc[school_summary_df["reading_score"] >= 70].groupby("school_name").count()["student_name"]

percentage_reading = (Total_passing_reading/student_count)*100

# overall percenatge pass 

Total_pass = school_summary_df.loc[(school_summary_df["math_score"] >= 70) & (school_summary_df["reading_score"] >= 70)].groupby("school_name").count()["student_name"]

##Total_pass = Total_passing_math & Total_passing_reading

percentage_total = (Total_pass/student_count)*100

#Create new dataframe for the school summary 

school_summary = pd.DataFrame({"School Type":School_type,
                               "Total Students":student_count,
                               "Total school budget":school_budget,
                               "Per student budget": student_budget,
                               "Average math score":average_math,
                               "Average reading score":average_reading,
                               "% passing math":percentage_math,
                               "% passing reading":percentage_reading,
                               "% overall passing":percentage_total})
#Display school summary

#Apply formatting

school_summary["Total school budget"] = school_summary["Total school budget"].map("${:,.0f}".format)
school_summary["Per student budget"] = school_summary["Per student budget"].map("${:,.2f}".format)

school_summary
```

Highest-Performing Schools (by % Overall Passing)

```python
# Highest-Performing Schools

Highest_performing_schools = school_summary.sort_values(by=["% overall passing"],ascending=False).head(5)
Highest_performing_schools
```

Math Scores by Grade

```python
# Use the code provided to separate the data by grade.
ninth_graders = school_data_complete.loc[school_data_complete["grade"] == "9th"]
tenth_graders = school_data_complete.loc[school_data_complete["grade"] == "10th"]
eleventh_graders = school_data_complete.loc[school_data_complete["grade"] == "11th"]
twelfth_graders = school_data_complete.loc[school_data_complete["grade"] == "12th"]


# Use the code to select only the `math_score`.
ninth_grade_math_scores = ninth_graders.groupby("school_name")["math_score"].mean()
tenth_grader_math_scores = tenth_graders.groupby("school_name")["math_score"].mean()
eleventh_grader_math_scores = eleventh_graders.groupby("school_name")["math_score"].mean()
twelfth_grader_math_scores = twelfth_graders.groupby("school_name")["math_score"].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({"9th": ninth_grade_math_scores, "10th":tenth_grader_math_scores,
                                     "11th":eleventh_grader_math_scores, "12th":twelfth_grader_math_scores})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

Reading Scores by Grade

```python
# Use the code provided to separate the data by grade.
ninth_grader_reading_scores = ninth_graders.groupby("school_name")["reading_score"].mean()
tenth_grader_reading_scores = tenth_graders.groupby("school_name")["reading_score"].mean()
eleventh_grader_reading_score = eleventh_graders.groupby("school_name")["reading_score"].mean()
twelfth_grader_reading_scores = twelfth_graders.groupby("school_name")["reading_score"].mean()


# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({"9th":ninth_grader_reading_scores , "10th":tenth_grader_reading_scores, 
                             "11th":eleventh_grader_reading_score, "12th":twelfth_grader_reading_scores})

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]

reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

Scores by School Spending

```python
# Copy school summary

school_spending_df = school_data_complete.copy()
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# pd.cut to categorize spending based on the bins.

school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"],
                                                               spending_bins,labels=labels)


spending_groups = school_data_complete.groupby("Spending Ranges (Per Student)")

#  Calculate averages for the desired columns. 
avg_math = spending_groups['math_score'].mean()
avg_read = spending_groups['reading_score'].mean()


#  Calculate pass marl for the desired columns. 
mark_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby("Spending Ranges (Per Student)")['Student ID'].count()/spending_groups['Student ID'].count()*100
mark_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby("Spending Ranges (Per Student)")['Student ID'].count()/spending_groups['Student ID'].count()*100
overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("Spending Ranges (Per Student)")["Student ID"].count()/spending_groups["Student ID"].count()*100

summary_table = pd.DataFrame({"Average Math Score": avg_math,
                              "Average Reading Score": avg_read,
                              "% Passing Math": mark_math,
                              "% Passing Reading": mark_read,
                              "% Overall Passing":overall 
                                    })
summary_table.index.name = "Spending Ranges (Per Student)"

summary_table = summary_table.reindex(labels)

# Display results
summary_table
```

Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
school_data_complete["size_bins"] = pd.cut(school_data_complete["size"], size_bins, labels = labels) 


#Categorize the spending based on the bins
by_bin = school_data_complete.groupby("size_bins")

# Calculate averages for the desired columns. 
avg_math = by_bin['math_score'].mean()

avg_read = by_bin['reading_score'].mean()

math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("size_bins")["Student ID"].count()/by_bin["Student ID"].count()*100
read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("size_bins")["Student ID"].count()/by_bin["Student ID"].count()*100
overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("size_bins")["Student ID"].count()/by_bin["Student ID"].count()*100

# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
size_summary = pd.DataFrame({"Average Math Score": avg_math,
                             "Average Reading Score": avg_read,
                             '% Passing Math': math,
                             '% Passing Reading': read,
                             "% Overall Passing": overall})

size_summary.index.name = "School Size"
size_summary = size_summary.reindex(labels)

# Display results and formating
size_summary.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
size_summary
```

```python
# Group the school_data_complete DataFrame by "School Type" and average the results.

school_type = school_data_complete.groupby("type")

avg_math_score = school_type["math_score"].mean()
avg_reading_score = school_type["math_score"].mean()
avg_passing_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("type")["Student ID"].count()/school_type["Student ID"].count()*100
avg_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("type")["Student ID"].count()/school_type["Student ID"].count()*100
avg_overall_passing = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("type")["Student ID"].count()/school_type["Student ID"].count()*100
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`
type_summary = pd.DataFrame({
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    '% Passing Math': avg_passing_math,
    '% Passing Reading': avg_passing_reading,
    "% Overall Passing": avg_overall_passing})

type_summary = type_summary[["Average Math Score", "Average Reading Score", '% Passing Math', '% Passing Reading', "% Overall Passing"]]
type_summary.index.name = "School Type"

#Formating
type_summary.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
type_summary
```
